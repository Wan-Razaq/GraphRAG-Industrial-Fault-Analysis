```python
from openai import OpenAI
import os

# Initialize client pointing to Cloudeka infrastructure 
client = OpenAI(
    base_url="https://dekallm.cloudeka.ai/v1",
    api_key=os.environ["DEKA_API_KEY"]
)

print("Sending request via OpenAI SDK...")

try:
    response = client.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {"role": "user", "content": "Who are you? How can you help me!"}
        ],
        temperature=0.7
    )

    print("\n[Connection Successful!]")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"\n[Error]: {e}")
```

```python
# Load environtment and repo root
from dotenv import load_dotenv
import sys
from pathlib import Path

load_dotenv()

repo_root = Path.cwd().parents[1]


#Append BAML to sys.path
sys.path.insert(0, str(repo_root / "src" / "models" / "baml"))
from baml_client.sync_client import b
from baml_client.types import FaultReport 

```

```python
# few-shot 20 cases for evaluation

import json
from pathlib import Path

# Path to the few-shot cases JSONL file
input_path = repo_root / "data" / "processed" / "20_cases_for_baml_fewshot.jsonl"

# Load 20 log_text strings
log_entries = []
with open(input_path, "r", encoding="utf-8") as f:
    for line in f:
        log_entries.append(json.loads(line))  # Each line is a raw log string

# Collect the results
extracted_results = []

for i, log_text in enumerate(log_entries):
    print(f"Processing case {i+1}/{len(log_entries)}")
    try:
        result: FaultReport = b.ExtractFaultInfo(log_text=log_text)
        extracted_results.append({
            "case_id": log_text.split("\n")[0].replace("Case-ID:", "").strip(),
            "result": result.model_dump()
        })
    except Exception as e:
        print(f" Failed to process case {i+1}: {e}")
        extracted_results.append({
            "case_id": log_text.split("\n")[0].replace("Case-ID:", "").strip(),
            "error": str(e)
        })

# Save the results to file
output_path = Path("baml_extracted_20_cases_gptoss.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(extracted_results, f, indent=2, ensure_ascii=False)

print(f" Extraction completed! Saved to: {output_path}")
```
